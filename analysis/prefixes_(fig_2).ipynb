```python
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
```

```python
import numpy as np
```

```python
cols = data.columns
cols
```

```python
data_ = data.copy()
for col in cols[2:]:
  data_['%s_diff' %col] = abs(data[col] - data['baseline'])

data_
```

```python
diff_cols = [i+'_diff' for i in cols[2:]]
data_diffs = data_[diff_cols]
data_diffs
```

```python
data_['avg_abs_diff'] = data_[diff_cols].mean(axis = 1)
data_
```

```python
diff_cols
```

```python
rqd_cols = ['model'] + list(diff_cols)  + ['avg_abs_diff']
rqd_cols
```

```python
ax = sns.heatmap(data_[rqd_cols].set_index('model'),
            center=0, cmap=sns.diverging_palette(220, 20, as_cmap=True))
ax.set_xlabel('Absolute Difference in Bias by Prefix')
ax.hlines([7], color = 'black', linestyles='dashed', *ax.get_xlim())
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, sharey = True, figsize = (17, 7.5))

sns.heatmap(data.set_index('model'), ax = ax[0],
            center=0, cmap=sns.diverging_palette(220, 20, as_cmap=True))
ax[0].set_xlabel('Bias by Prefix')
ax[0].hlines([7], color = 'black', linestyles='dashed', *ax[0].get_xlim())
ax[0].set_title('a) Political Bias due to Prefixes')

sns.heatmap(data_[rqd_cols].set_index('model'), ax = ax[1],
            center=0, cmap=sns.diverging_palette(220, 20, as_cmap=True))
ax[1].set_xlabel('Absolute Difference in Bias by Prefix')
ax[1].hlines([7], color = 'black', linestyles='dashed', *ax[1].get_xlim())
ax[1].set_title('b) Absolute Difference in Political Bias\nfrom Baseline due to Prefixes')

plt.tight_layout()
plt.savefig('prefix.pdf')
```
